# Iteración de punto fijo

_**Modo de uso**_: una vez seleccionada la celda que contiene el código, pulsar May+Intro

```python
#################
# Importaciones #
#################

import numpy as np
import pandas as pd
import sympy as sp
from IPython.display import Markdown, Math, Latex

###########################################
# Declaración de algunas variables útiles #
###########################################

p0_matrix = []
fx_matrix = []
gp_matrix = []
precision_matrix = []
d1col_matrix = []
x = sp.symbols('x')

###############################################
# Se define la función que aplicará el método #
###############################################

def puntofijo(g, p0, tol, iter):
    i = 1
    d1col_matrix.append(p0)
    
    while i <= iter :
        p = g(p0)
        p0_matrix.append(p0)
        gp_matrix.append(p)
        precision_matrix.append(np.abs(p-p0))
        d1col_matrix.append(p)
        if np.abs(p-p0) >= tol :
            i += 1
            p0 = p
            #return (p, i)
        else :
            break
            
    matrix_long = len(gp_matrix)-1
    if np.abs(gp_matrix[matrix_long]-p0_matrix[matrix_long]) < tol:
        return (p, matrix_long + 1)
    else:
        return ("*El método ha fallado después de **{}** iteraciones*\n".format(len(gp_matrix)), -1)
    
##################################################################
# Recolección de datos a través de la interacción con el usuario #
##################################################################

gfunc = input("Introduzca g(x): ")
pinicial_float = eval(input("Introduzca p0: "))
tol_float = eval(input("Introduzca la tolerancia: "))
iter_int = eval(input("Introduzca el número de iteraciones: "))

gx = sp.lambdify(x, gfunc)

res, niter = puntofijo(gx, pinicial_float, tol_float, iter_int) # Llamada a la función

##############################
# Se muestran los resultados #
##############################

#print ("\n\nResultado:\n---------> Raíz: ", res, "con punto inicial: ", pinicial_float, " y tolerancia: ", tol_float, " después de ", niter, " iteraciones")
if niter == -1:
    display(Markdown('\n\nResultado: {}'.format(res)))
else:    
    display(Markdown('\n\nResultado para la función: $f(x) = {}$\n\nRaíz: **{}**, con punto inicial **{}**  y tolerancia **{}** después de **{}** iteraciones\n'.format(gfunc, res, pinicial_float, tol_float, niter)))
 
    pd.set_option('display.precision', 10)
    df = pd.DataFrame({'$p_0$':p0_matrix,
                      '$g(p_0)$' : gp_matrix,
                      '$Error$': precision_matrix})
    df.index = np.arange(1, niter+1, 1)
    df.index.name = '$n$'

    datos_1col = pd.DataFrame({'$p_n$': d1col_matrix})
    datos_1col.index.name ='$n$'

    display(datos_1col)

    display(df)

    pndf = pd.DataFrame({'pn': p0_matrix})
    pn1df = pd.DataFrame({'pn1': gp_matrix})

#####################################################
# Los datos tabulados se guardan como archivo Latex # 
#####################################################

    df.to_latex("puntofijo_general.tex")
    datos_1col.to_latex("puntofijo_1col.tex")

#####################################################
# Los datos tabulados se guardan como archivo CSV   # 
#####################################################

    datos_1col.to_csv("puntofijo.csv", index=None, header=False)
    pndf.to_csv('puntofijo-p_n.csv', index=None, header=False)
    pn1df.to_csv('puntofijo-p_n1.csv', index=None, header=False)
```
