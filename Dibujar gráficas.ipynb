# Apartado b)

```python
import numpy as np
from pylab import *
from matplotlib import pyplot as plt
import math
import sympy as sp
import warnings
warnings.filterwarnings("ignore")
#plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'lines.linewidth': 1})
x1 = [1, 2]
x2 = [0.2, 0.2]    
ys = [0.2, -12.5]

x=np.arange(0.07,4.0,0.01)
x2 = np.arange(0.13, 4, 0.01)
xrectay = np.arange(0, 4, 0.01)
y=1/(2*x**2)-5/x
mangular = 1/(2*x2**2)
potencial = -5/x2
rectay = -12.5*xrectay/xrectay
fig=plt.figure(figsize=(7, 5), dpi= 80, facecolor='w', edgecolor='k')

ax=fig.add_axes([0,0,1,1])
plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
#plt.title('Diagrama de energía')

#ax.grid(True)
plt.xlabel('r', fontsize=20)
plt.ylabel('$V_{ef}(r)$', fontsize=20)
ax.plot(x,y, label='$V_{ef}=\\frac{1}{2r^2}-\\frac{5}{r}$')
ax.plot(x2, potencial, '-.', label='$V(r)=-\\frac{5}{r}$')
ax.plot(x2, mangular, '-.', label='$\\frac{1}{2r^2}$')
#ax.plot(0.2, -12.5,"--", label='$r_{min}$')
ax.plot([0.2, 0.2], [0.2, -12.5], ":", label='$r_{min}$')
ax.plot(xrectay,rectay,"--", linewidth=1, label='$E=V_{ef}(r_{min})$')
#ax.plot(x2, ys, 'black', linestyle='--', marker='')
ax.legend(loc='uper right')
ax.annotate('$r_{min}$', [0.2, 0.2])
ax.annotate('$E(r_{min})$', [1, -12]) # Energía total
ax.annotate('$V(r)=-\\frac{5}{r}$', [0.2, -30])
ax.annotate('$\\frac{1}{2r^2}$', [0.2, 20])

show()

fig.savefig('apartado-b.png',bbox_inches='tight')
```
