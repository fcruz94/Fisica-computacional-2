```python
import numpy as np
import pandas as pd
from IPython.display import display, Markdown
import sympy as sp
from sympy import init_session


#init_session(use_latex=True)

def u_cal(u, n): 

    temp = u; 
    for i in range(1, n): 
        temp = temp * (u - i); 
    return temp; 

# calculating factorial of given number n 
def factorial(n): 
    f = 1; 
    for i in range(2, n + 1): 
        f *= i; 
    return f; 

def newton_prog(x, array, fxarray):
    """
    Implementaci√≥n del m√©todo de diferencias progresivas de Newton.
    
    Argumentos:
        float x: punto a interpolar
        array array: matriz conteniendo los x_i valores
        array fxarray: matriz conteniendo los f(x_i) valores
    Devuelve:
        array parray
    """
    global xarray
    global nfx
    global parray
    
    xarray = np.asarray(array) # Contiene los n√∫meros  x_0, ...x_n
    fxarray = np. asarray(fxarray) # Contiene los valores f(x_0), ...f(x_n)
       
    nx = len(xarray)
    nfx = len(fxarray)
    parray = np.zeros(((nfx), (nfx)))
    
    for i in range(0, nfx):
        parray[i, 0] = fxarray[i]
    #print(parray)
    
    for i in range(1, nfx): 
        for j in range(nfx - i): 
            parray[j, i] = parray[j + 1, i - 1] - parray[j, i - 1]
            
#     for i in range(nfx): 
#         print(xarray[i], end = "\t"); 
#         for j in range(nfx - i): 
#             print(parray[i, j], end = "\t")
#         print("");  
    return(parray)

# Ejercicio b.21)
# display(Markdown('__Ejercicio b.2.1)__'))
# xa = [0.82, 0.84, 0.86, 0.88]
# ya = [1650, 1450, 1250, 1050]
# x= 0.8705

# Ejercicio b.22)
display(Markdown('__Ejercicio b.2.2)__'))
xa = [0.84, 0.86, 0.88, 0.90]
ya = [1450, 1250, 1050, 870]
x= 0.8705

# xa = [1.0, 1.3, 1.6, 1.9, 2.2]
# ya = [0.7651977, 0.62008860, 0.4554022, 0.2818186, 0.1103623]
# x = 1.5
# xa = [-1/2, 0, 1/2, 1]
# ya = [-4, 3, 13/2, 8]
# x = 0.5
# Llamada a la funci√≥n
p = newton_prog(x, xa, ya)


sum = parray[0, 0]; 
u = (x - xarray[0]) / (xarray[1] - xarray[0]); 
for i in range(1,nfx): 
    sum +=  (u_cal(u, i) * parray[0][i]) / factorial(i); 

pd.set_option('display.precision', 7)
columnas = ['$f(x_i)$', '$\\Delta ^1 f_i$', '$\\Delta ^2 f_i$', '$\\Delta ^3 f_i$']
tabla = pd.DataFrame(p, columns=columnas)
tabla.insert(loc=0, column='$x_i$', value=xa)

# tabla.to_csv("progresiva_newton-b21.csv", index=None, header=False)
# tabla.to_latex('progresiva_newton-b21.tex')

display(tabla)

display(Markdown('El resultado interpolado es: $f({})= {}$'.format(x, round( sum, 0))))

```

el polinomio en s es: 

P(s)=-4+7s/1!-3.5*[s+(s-1)/!]+1.5+[s*(s-1)+(s-2)/3!]. 

Para ponerlo en funci√≥n de x hay que tener en cuenta que s=(x-x_0)/h donde h es la diferencia entre los x_i, en este caso ser√≠a 1/2 de modo que:

s=(x+1/2)/1/2 = 2x+1

sustituyendo en el polinomio se obtiene:

P(x)= = 2x^3 ‚àí 7x^2 + 10^x + 3.
