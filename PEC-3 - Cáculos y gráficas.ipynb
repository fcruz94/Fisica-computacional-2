```python
"""
Derivada de la función de error
"""

abscisa = arange(0.82, 0.88, 0.0001)
abscisa2 = arange(0.84, 0.90,0.0001)
puntos1 = [0.82, 0.84, 0.86, 0.88,]
puntos2 = [0.84, 0.86, 0.88, 0.90]

# simbólica:
funcions = -ln(x)*(5730/ln(2))
# Sustituyendo los valores numéricos
# funcion = -ln(x)*(5730/0.6931472)
poli = (x-0.82)*(x-0.84)*(x-0.86)*(x-0.88)

display(Markdown("Función de antigüedad $t$: "),funcions)
display(Markdown('Funcion/fórmula de error: '))
display(Markdown('$E=\\frac{f^{n+1}\\xi}{(n+1!)}\\prod(x-x_i$)'))
pcriticos = solve(diff((x-0.82)*(x-0.84)*(x-0.86)*(x-0.88)))
display(Markdown('Puntos críticos del factor polinomio (ver gráfica al final): '), pcriticos )

# display(Markdown("Numérica: "),funcion)
for i in range(4):
#     funcion = diff(funcion, x)
    funcions = diff(funcions, x)


display(Markdown("Derivada $t^{4}$: "),funcions)
#isplay(Markdown("Derivada 4 (Numérica): "),funcion)

xi_factor = funcions*(1/factorial(4))

error_func = xi_factor*(x-0.82)*(x-0.84)*(x-0.86)*(x-0.88)
display(Markdown('Funcion del error: '), error_func)

display(Markdown('Valor del factor $\\xi$ en 0.82'), xi_factor.evalf(subs={x:0.82}))
display(Markdown('Valor $\\xi$ en 0.88'), xi_factor.evalf(subs={x:0.88}))

xi_maximo = xi_factor.evalf(subs={x:0.82})
xi_minimo = xi_factor.evalf(subs={x:0.88})
# el máximo es para 0.82 y los (x-x_i) tienen un máximo en 0.85
error_max = xi_factor.evalf(subs={x:0.82})*(0.85-0.82)*(0.85-0.84)*(0.85-0.86)*(0.85-0.88)
display(Markdown('aqui: '), xi_minimo*(0.85-0.82)*(0.85-0.84)*(0.85-0.86)*(0.85-0.88))
# error_max = error_func.evalf(subs={x:0.82})
#display(Markdown('Error máximo: '), error_max)
display(Markdown('Error máximo, tomando el valor mayor del factor $\\xi$ y el máximo del factor polinomio: '), xi_maximo*poli.evalf(subs={x:pcriticos[1]}))
display(Markdown('Error mínimo, tomando el valor menor del factor $\\xi$ y el primer mínimo del factor polinomio: '), xi_minimo*poli.evalf(subs={x:pcriticos[0]}))
display(Markdown('Error mínimo, tomando el valor menor del factor $\\xi$ y el segundo mínimo del factor polinomio: '), xi_minimo*poli.evalf(subs={x:pcriticos[2]}))

"""
GRAFICAS
"""


fig1=plt.figure(figsize=(7, 7), dpi= 80, facecolor='w', edgecolor='k')
fig2=plt.figure(figsize=(8, 7), dpi= 80, facecolor='w', edgecolor='k')
fig4 = plt.figure( dpi = 80, facecolor='w', edgecolor='k')

ax1=fig1.add_axes([0,0,1,1])
ax1.grid(True)

ax2=fig2.add_axes([0,0,1,1])
ax2.grid(True)
ax2.set_xlabel('$N_{14C}$', fontsize=18)
ax2.set_ylabel('Error', fontsize=18)
ax4=fig4.add_axes([0,0,1,1])
ax4.grid(True)
# ax3.set_xlabel('$N_{14C}$', fontsize=18)

peval= 0.8705
# Grafíca de los puntos críticos del polinomio
ax1.plot(abscisa, [poli.evalf(subs={x:i}) for i in abscisa], label='$P(x)=(x-0.82)(x-0.84)(x-0.86)(x-0.88)$')
ax1.plot(pcriticos, [poli.evalf(subs={x:i}) for i in pcriticos], 'o', label='Puntos críticos')
ax1.legend(loc="lower center")
fig1.savefig('pec-3/puntoscriticos.png',bbox_inches='tight')
# Gráfica de la función de error uitilizando los valores de xi que la maximizan y la minimizan apartado b1
# Maximiza
ax2.plot(abscisa, [xi_maximo*poli.evalf(subs={x:i}) for i in abscisa], label='$E=\\frac{f^{n+1}\\xi}{(n+1)!}\\prod(x-x_i)$ (Valor máximo)')
ax2.plot(peval, xi_maximo*poli.evalf(subs={x:peval}), 'or', label="$N_{14c}=0.8705$")
# Minimiza
ax2.plot(abscisa, [xi_minimo*poli.evalf(subs={x:i}) for i in abscisa], label='$E=\\frac{f^{n+1}\\xi}{(n+1)!}\\prod(x-x_i)$ (Valor mínimo)')
ax2.plot(peval, xi_minimo*poli.evalf(subs={x:peval}), 'or', label="$N_{14c}=0.8705$")
ax2.legend(loc="best", fontsize=14)
ax2.set_title('Función de error utilizando el valor máximo/mínimo del factor $\\xi$')
fig2.savefig('pec-3/funcion_error_max_min.png',bbox_inches='tight')

ax4.plot(abscisa, [xi_factor.evalf(subs={x:i}) for i in abscisa], label='$\\Xi=\\frac{34380}{\\ln(2)(N_{14C})^4}$')
ax4.plot(0.82, xi_factor.evalf(subs={x:0.82}), 'or', label='$N_{14C}=0,82$')
ax4.plot(0.88, xi_factor.evalf(subs={x:0.88}), 'og', label='$N_{14C}=0,88$')
ax4.legend(loc="best", fontsize=14)
# fig4.savefig('pec-3/xi_mayuscula_082-088.png',bbox_inches='tight')

# ferror= (2865*(0.8705-0.88)*(0.8705-0.86)*(0.8705-0.84)*(0.8705-0.82))/(2*x**4*log(2))
# display(Markdown("Version Fran: "), ferror)
```
